```python
# Tabela exemplo 21

# Função
def f(x):
    return x * math.log(x) - 1

# Derivada da Função
def df(x):
    return math.log(x) + 1

# φ(x)
def phi(x):
    return x - 1.3 * (x * math.log(x) - 1)

#Parâmetros
eps1    = 1e-7
eps2    = 1e-7
a       = 2
b       = 3
x0MPF   = 2.5
x0N     = 2.5
x0S     = 2.3
x1      = 2.7
```

```python
# Métodos iterativos

def bissecao(f, a, b, eps1, eps2, max_iter=1000):
    """
    Implementação do método da bisseção.
    """

    inicio = time.perf_counter()

    # Verificação inicial
    if f(a) * f(b) >= 0:
        raise ValueError("É necessário que f(a)*f(b) < 0")

    # Passo 3
    k = 1

    while k <= max_iter:

        # Passo 2
        # Critério de parada por tamanho do intervalo
        if (b - a) < eps1:
            x_barra = (a + b) / 2
            erro = (b - a) / 2
            fim = time.perf_counter()
            return {
                "raiz": x_barra,
                "erro": erro,
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # Critério de parada por valor da função
        if abs(f(a)) < eps2:
            erro = (b - a) / 2
            fim = time.perf_counter()
            return {
                "raiz": a,
                "erro": erro,
                "iteracoes": k,
                "tempo": fim - inicio
            }

        if abs(f(b)) < eps2:
            erro = (b - a) / 2
            fim = time.perf_counter()
            return {
                "raiz": b,
                "erro": erro,
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # Passo 4
        M = f(a)

        # Passo 5 (bisseção)
        x = (a + b) / 2

        # Passo 6
        if abs(f(x)) < eps2:
            erro = (b - a) / 2
            fim = time.perf_counter()
            return {
                "raiz": x,
                "erro": erro,
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # Passo 7 e 8
        if M * f(x) > 0:
            a = x
        else:
            b = x

        k += 1

    fim = time.perf_counter()
    raise RuntimeError(
        f"Número máximo de iterações atingido ({max_iter}) "
        f"em {fim - inicio:.6f} segundos"
    )


def posicao_falsa(f, a, b, eps1, eps2, max_iter=1000):
    """
    Método da Posição Falsa
    """

    # =========================
    # Passo 1 – Dados iniciais
    # =========================
    inicio = time.perf_counter()

    if f(a) * f(b) >= 0:
        raise ValueError("É necessário que f(a)*f(b) < 0")

    # ==========
    # Passo 3
    # ==========
    k = 1

    while k <= max_iter:

        # ==========================================
        # Passo 2 – Critérios iniciais de parada
        # ==========================================
        if (b - a) < eps1:
            x_barra = (a + b) / 2
            erro = (b - a) / 2
            fim = time.perf_counter()
            return {
                "raiz": x_barra,
                "erro": erro,
                "iteracoes": k,
                "tempo": fim - inicio
            }

        if abs(f(a)) < eps2:
            erro = (b - a) / 2
            fim = time.perf_counter()
            return {
                "raiz": a,
                "erro": erro,
                "iteracoes": k,
                "tempo": fim - inicio
            }

        if abs(f(b)) < eps2:
            erro = (b - a) / 2
            fim = time.perf_counter()
            return {
                "raiz": b,
                "erro": erro,
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # ==========
        # Passo 4
        # ==========
        M = f(a)

        # ==========================================
        # Passo 5 – Fórmula da posição falsa
        # x = (a f(b) - b f(a)) / (f(b) - f(a))
        # ==========================================
        x = (a * f(b) - b * f(a)) / (f(b) - f(a))

        # ==========
        # Passo 6
        # ==========
        if abs(f(x)) < eps2:
            erro = abs(b - a) / 2
            fim = time.perf_counter()
            return {
                "raiz": x,
                "erro": erro,
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # ======================================
        # Passo 7 – Atualização do intervalo
        # ======================================
        if M * f(x) > 0:
            a = x
        else:
            # ==========
            # Passo 8
            # ==========
            b = x

        # ==========
        # Passo 9
        # ==========
        if (b - a) < eps1:
            x_barra = (a + b) / 2
            erro = (b - a) / 2
            fim = time.perf_counter()
            return {
                "raiz": x_barra,
                "erro": erro,
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # ==========
        # Passo 10
        # ==========
        k += 1

    fim = time.perf_counter()
    raise RuntimeError(
        f"Número máximo de iterações atingido ({max_iter}) "
        f"em {fim - inicio:.6f} segundos"
    )


def ponto_fixo(f, phi, x0, eps1, eps2, max_iter=1000):
    """
    Método do Ponto Fixo.

    f   : função original f(x)
    phi : função de iteração φ(x)
    x0  : aproximação inicial
    """

    inicio = time.perf_counter()

    # =====================================
    # Passo 2 – Teste inicial de convergência
    # =====================================
    if abs(f(x0)) < eps1:
        fim = time.perf_counter()
        return {
            "raiz": x0,
            "erro": 0.0,
            "iteracoes": 0,
            "tempo": fim - inicio
        }

    # ==========
    # Passo 3
    # ==========
    k = 1

    while k <= max_iter:

        # ==========
        # Passo 4
        # x1 = φ(x0)
        # ==========
        x1 = phi(x0)

        # ==========================================
        # Passo 5 – Critérios de parada
        # |f(x1)| < ε1  ou  |x1 - x0| < ε2
        # ==========================================
        if abs(f(x1)) < eps1 or abs(x1 - x0) < eps2:
            fim = time.perf_counter()
            return {
                "raiz": x1,
                "erro": abs(x1 - x0),
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # ==========
        # Passo 6
        # ==========
        x0 = x1

        # ==========
        # Passo 7
        # ==========
        k += 1

    fim = time.perf_counter()
    raise RuntimeError(
        f"Número máximo de iterações atingido ({max_iter}) "
        f"em {fim - inicio:.6f} segundos"
    )


def newton(f, df, x0, eps1, eps2, max_iter=1000):
    """
    Método de Newton.

    f  : função f(x)
    df : derivada f'(x)
    x0 : aproximação inicial
    """

    inicio = time.perf_counter()

    # =========================
    # Passo 2
    # Se |f(x0)| < ε1, faça x̄ = x0. FIM.
    # =========================
    if abs(f(x0)) < eps1:
        fim = time.perf_counter()
        return {
            "raiz": x0,
            "erro": 0.0,
            "iteracoes": 0,
            "tempo": fim - inicio
        }

    # =========================
    # Passo 3
    # =========================
    k = 1

    while k <= max_iter:

        # =========================
        # Passo 4
        # x1 = x0 - f(x0) / f'(x0)
        # =========================
        df_x0 = df(x0)
        if df_x0 == 0:
            raise ZeroDivisionError("Derivada nula. Método de Newton falhou.")

        x1 = x0 - f(x0) / df_x0

        # =========================
        # Passo 5
        # Se |f(x1)| < ε1 ou |x1 - x0| < ε2
        # =========================
        if abs(f(x1)) < eps1 or abs(x1 - x0) < eps2:
            fim = time.perf_counter()
            return {
                "raiz": x1,
                "erro": abs(x1 - x0),
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # =========================
        # Passo 6
        # =========================
        x0 = x1

        # =========================
        # Passo 7
        # =========================
        k += 1

    fim = time.perf_counter()
    raise RuntimeError(
        f"Número máximo de iterações atingido ({max_iter}) "
        f"em {fim - inicio:.6f} segundos"
    )


def metodo_secante(f, x0, x1, eps1, eps2, max_iter=1000):
    """
    Implementação do método da secante
    """
    inicio = time.perf_counter()

    # Passo 1: Dados iniciais (já recebidos como parâmetros)

    # Passo 2
    if abs(f(x0)) < eps1:
        fim = time.perf_counter()
        return {
            "raiz": x0,
            "erro": abs(f(x0)),
            "iteracoes": 0,
            "tempo": fim - inicio
        }

    # Passo 3
    if abs(f(x1)) < eps1 or abs(x1 - x0) < eps2:
        fim = time.perf_counter()
        return {
            "raiz": x1,
            "erro": abs(f(x1)),
            "iteracoes": 0,
            "tempo": fim - inicio
        }

    # Passo 4
    k = 1

    while k <= max_iter:
        # Passo 5: Fórmula do método da secante
        f_x0 = f(x0)
        f_x1 = f(x1)
        denominador = f_x1 - f_x0

        # Evitar divisão por zero
        if abs(denominador) < 1e-15:
            break

        x2 = x1 - (f_x1 / denominador) * (x1 - x0)

        # Passo 6
        if abs(f(x2)) < eps1 or abs(x2 - x1) < eps2:
            fim = time.perf_counter()
            return {
                "raiz": x2,
                "erro": abs(f(x2)),
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # Passo 7
        x0 = x1
        x1 = x2

        # Passo 8
        k += 1

    # Se atingiu o máximo de iterações sem convergir
    fim = time.perf_counter()
    return {
        "raiz": x1,
        "erro": abs(f(x1)),
        "iteracoes": k-1,
        "tempo": fim - inicio,
        "mensagem": "Máximo de iterações atingido"
    }
```

```python
# Exemplo 1

# Função
def f(x):
    return x**5 - 3.7*x**4 + 7.4*x**3 - 10.8*x**2 + 10.8*x - 6.8

def df(x):
    return 5*x**4 - 14.8*x**3 + 22.2*x**2 - 21.6*x + 10.8

def phi(x):
    return x - f(x) / df(x)

# Lista de coeficientes
coef = [-6.8, 10.8, -10.8, 7.4, -3.7, 1]

#Intervalo
#[1, 2]

#Parâmetros
eps1    = 1e-4
eps2    = 1e-4
x0MPF   = 1.5
x0S     = 1.2
x1      = 1.4
```

```python

def metodo_letra_b(coef, x0, eps1, eps2, max_iter=1000):
    """
    Método do pseudocódico da letra B.
    """

    inicio = time.perf_counter()

    n = len(coef) - 1   # grau do polinômio

    # =========================
    # Passo 1
    # =========================
    x = x0
    deltax = x

    # =========================
    # Passo 2
    # =========================
    for k in range(1, max_iter + 1):

        # b = a_n
        b = coef[n]
        c = b

        # Para i = n-1, ..., 1
        for i in range(n - 1, 0, -1):
            b = coef[i] + b * x
            c = b + c * x

        # b = a_0 + b x
        b = coef[0] + b * x

        # Se |b| ≤ ε1 → convergiu (p(x) ≈ 0)
        if abs(b) <= eps1:
            fim = time.perf_counter()
            return {
                "raiz": x,
                "erro": abs(deltax),
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # deltax = b / c
        deltax = b / c

        # x = x − deltax
        x = x - deltax

        # Se |deltax| ≤ ε2 → convergiu
        if abs(deltax) <= eps2:
            fim = time.perf_counter()
            return {
                "raiz": x,
                "erro": abs(deltax),
                "iteracoes": k,
                "tempo": fim - inicio
            }

    # =========================
    # Passo 3
    # =========================
    fim = time.perf_counter()
    return {
        "raiz": None,
        "erro": None,
        "iteracoes": max_iter,
        "tempo": fim - inicio,
        "mensagem": "Não houve convergência com max_iter iterações"
    }
```

```python
resultado = metodo_letra_b(
    coef,
    x0MPF,
    eps1,
    eps2
)

print(f"Raiz aproximada : {resultado['raiz']}")
print(f"Erro            : {resultado['erro']}")
print(f"Iterações       : {resultado['iteracoes']}")
print(f"Tempo (s)       : {resultado['tempo']:.8f}")
print(f"F(Raiz)         : {f(resultado['raiz'])}")
```

```python

def letra_b_mpf(f, phi, x0, eps1, eps2, max_iter=1000):
    """
    Método do ponto fixo adaptado a partir do pseudocódigo da letra B.
    """

    inicio = time.perf_counter()

    # =========================
    # Passo 1 – Aproximação inicial
    # =========================
    x = x0
    # Incicializa com o maior número disponível no python
    deltax = float("inf")

    # =========================
    # Passo 2 – Iterações
    # =========================
    for k in range(1, max_iter + 1):

      # =========================
      # Atualização do ponto fixo
      # =========================
      x_novo = phi(x)

      # =========================
      # Diferença entre iterações
      # =========================
      deltax = x_novo - x

      # Atualiza x
      x = x_novo

      # ==========================================
      # Passo 5 – Critérios de parada
      # |f(x)| < ε1  ou  |x_k − x_{k-1}| < ε2
      # ==========================================
      if abs(f(x)) < eps1 or abs(deltax) < eps2:
          fim = time.perf_counter()
          return {
              "raiz": x,
              "erro": abs(deltax),
              "iteracoes": k,
              "tempo": fim - inicio
          }

    # =========================
    # Passo 3 – Não convergiu
    # =========================
    fim = time.perf_counter()
    return {
        "raiz": None,
        "erro": None,
        "iteracoes": max_iter,
        "tempo": fim - inicio,
        "mensagem": "Não houve convergência com max_iter iterações"
    }

```

```python
resultado = letra_b_mpf(f, phi, x0MPF, eps1, eps2)

print(f"Raiz aproximada : {resultado['raiz']}")
print(f"Erro            : {resultado['erro']}")
print(f"Iterações       : {resultado['iteracoes']}")
print(f"Tempo (s)       : {resultado['tempo']:.8f}")
print(f"F(Raiz)         : {f(resultado['raiz'])}")
```

```python

def letra_b_secante(coef, x0, x1, eps1, eps2, max_iter=1000):
    """
    Método da Secante adaptado a partir do pseudocódigo da letra B.
    """

    inicio = time.perf_counter()

    # =========================
    # Passo 1 – Aproximações iniciais
    # =========================
    x_prev = x0     # x_{k-1}
    x = x1          # x_k
    deltax = float("inf")

    # Avalia f(x_{k-1}) e f(x_k)
    f_prev = avalia_polinomio(coef, x_prev)
    f_x = avalia_polinomio(coef, x)

    # =========================
    # Passo 2 – Iterações
    # =========================
    for k in range(1, max_iter + 1):

        # =========================================
        # Teste |f(x_k)| ≤ ε1
        # (mesmo critério da letra B)
        # =========================================
        if abs(f_x) <= eps1:
            fim = time.perf_counter()
            return {
                "raiz": x,
                "erro": abs(deltax),
                "iteracoes": k,
                "tempo": fim - inicio
            }

        # =========================================
        # Fórmula da secante
        # x_{k+1} = x_k − f(x_k) * (x_k − x_{k−1}) /
        #            (f(x_k) − f(x_{k−1}))
        # =========================================
        denominador = f_x - f_prev
        if denominador == 0:
            break  # evita divisão por zero

        x_novo = x - f_x * (x - x_prev) / denominador

        # Diferença entre iterações
        deltax = x_novo - x

        # Atualiza os valores para a próxima iteração
        x_prev = x
        f_prev = f_x

        x = x_novo
        f_x = avalia_polinomio(coef, x)

        # =========================================
        # Teste |Δx| ≤ ε2
        # =========================================
        if abs(deltax) <= eps2:
            fim = time.perf_counter()
            return {
                "raiz": x,
                "erro": abs(deltax),
                "iteracoes": k,
                "tempo": fim - inicio
            }

    # =========================
    # Passo 3 – Não convergiu
    # =========================
    fim = time.perf_counter()
    return {
        "raiz": None,
        "erro": None,
        "iteracoes": max_iter,
        "tempo": fim - inicio,
        "mensagem": "Não houve convergência com max_iter iterações"
    }

def avalia_polinomio(coef, x):
    """
    Avalia p(x) pelo método de Horner
    """
    resultado = coef[-1]
    for a in reversed(coef[:-1]):
        resultado = a + resultado * x
    return resultado

```

```python
resultado = letra_b_secante(coef, x0S, x1, eps1, eps2)

print(f"Raiz aproximada : {resultado['raiz']}")
print(f"Erro            : {resultado['erro']}")
print(f"Iterações       : {resultado['iteracoes']}")
print(f"Tempo (s)       : {resultado['tempo']:.8f}")
print(f"F(Raiz)         : {f(resultado['raiz'])}")
```
